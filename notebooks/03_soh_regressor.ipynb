# Phase 3 SOH Regressor Diagnostics

This notebook inspects the SOH models trained by `scripts/train_soh.py`.

```python
import pandas as pd
import matplotlib.pyplot as plt
from triagenet.config import REPO_ROOT, DATA_PROCESSED
report_dir = REPO_ROOT / 'reports' / 'phase3_soh'
metrics = pd.read_csv(report_dir / 'metrics.csv')
display(metrics)
```

```python
pred = pd.read_csv(report_dir / 'ensemble_gpr_predictions.csv')
display(
    pred.assign(abs_error=(pred.soh_mean - pred.soh_true).abs())
    .sort_values('abs_error', ascending=False)
    .head()
)
```

```python
for image in sorted(report_dir.glob('*_scatter_interval.png'))[:6]:
    plt.figure(figsize=(6, 4))
    plt.imshow(plt.imread(image))
    plt.title(image.name)
    plt.axis('off')
    plt.show()
```

## What I observed

The model predicts SOH without direct capacity features, so the useful diagnostic is calibration: whether the 90% intervals contain roughly 90% of held-out measured SOH. Any disagreement cases above five percentage points should be inspected as possible partial-discharge, temperature, or protocol artifacts.
